```python
#information on total number of rows,columns,data type of each column and memory requirement
dataset.info()
```

```python
#Number of missing values in each column
dataset.isnull().sum()
```

```python
#Dropping the null values from  listed columns
columns=['Region','IncomeGroup']
dataset=dataset.dropna(subset=columns)
```

```python
#Displaying the dataframe
dataset
```

```python
#Number of missing values in each column after dropping the null valued rows
dataset.isnull().sum()
```

```python
#Grouping the data based on region and income group
group_data=dataset.groupby(['Region','IncomeGroup']).size()
group_data
```

```python
##Grouping the data based on region and income group using unstack()
group_data1=dataset.groupby(['Region','IncomeGroup']).size().unstack(fill_value=0)
group_data1
```

```python
#Barchart represention of grouped data
group_data1.plot(kind='bar',figsize=(10,8))
plt.xlabel('Region')
plt.ylabel('Number of countries')
plt.title('Income distribution in various regions')
plt.legend(title='Income group')
```

```python
# Stacked Barchart represention of grouped data
group_data1.plot(kind='bar',stacked=True,figsize=(10,8))
plt.xlabel('Region')
plt.ylabel('Number of countries')
plt.title('Income distribution in various regions')
plt.legend(title='Income group')
```

```python
#Countplot to represent the number of countries in each region
plt.figure(figsize=(20,20))
sns.countplot(x='Region', data=dataset)
plt.title("Number of countries in each region")
plt.show()
```
